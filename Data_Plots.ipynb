```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


input_data = "Resources/Clean_Data.csv"
```

```python
main_df= pd.read_csv(input_data)
main_df.sort_values(["year","Happiness score"])
```

```python

#main_df_sorted=main_df.sort_values(["Country","year"]).reset_index()
#main_df_sorted.head()

pivot_df= main_df.pivot(index='year', columns='Country', values='Happiness score')
pivot_df

x_axis=[2014,2015,2016,2017]

column_names=pivot_df.columns
#print(column_names)
for cntry in column_names:
    y_axis=pivot_df[cntry]
    plt.plot(x_axis,y_axis)

#plt.legend()
plt.figure(figsize=(16,7))
plt.show()
```

```python

country_lim=7
top_bottom_df_2014=pd.concat([df_2014.nsmallest(country_lim,"Happiness score"),df_2014.nlargest(country_lim,"Happiness score").sort_values("Happiness score")])
top_bottom_df_2015=pd.concat([df_2015.nsmallest(country_lim,"Happiness score"),df_2015.nlargest(country_lim,"Happiness score").sort_values("Happiness score")])
top_bottom_df_2016=pd.concat([df_2016.nsmallest(country_lim,"Happiness score"),df_2016.nlargest(country_lim,"Happiness score").sort_values("Happiness score")])
top_bottom_df_2017=pd.concat([df_2017.nsmallest(country_lim,"Happiness score"),df_2017.nlargest(country_lim,"Happiness score").sort_values("Happiness score")])
```

```python
x_axis_2014 = np.arange(len(top_bottom_df_2014["Happiness score"]))


plt.figure(figsize=(country_lim*2.5,4))
for i, v in enumerate(top_bottom_df_2014["Happiness score"]):
    plt.text(x_axis_2014[i] , v + 0.025, str(round(v,1)))

plt.title("2014 Happiest/Least Happiest")
plt.ylabel(" Happiness Rating")
plt.xticks(x_axis_2014, top_bottom_df_2014["Country"])
plt.bar(x_axis_2014, top_bottom_df_2014["Happiness score"], color='r', alpha=0.5, align="center")

```

```python
x_axis_2015 = np.arange(len(top_bottom_df_2015["Happiness score"]))

plt.figure(figsize=(country_lim*2.5,4))
for i, v in enumerate(top_bottom_df_2015["Happiness score"]):
    plt.text(x_axis_2015[i] , v + 0.025, str(round(v,1)))

plt.title("2015 Happiest/Least Happiest")
plt.ylabel(" Happiness Rating")
plt.xticks(x_axis_2015, [l.replace(" ","\n") for l in top_bottom_df_2015["Country"].tolist()])
plt.bar(x_axis_2015, top_bottom_df_2015["Happiness score"], color='r', alpha=0.5, align="center")
```
