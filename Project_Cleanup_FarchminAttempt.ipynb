```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
file_2018 = "Resources/WHR2018Chapter2OnlineData_2018.xls"
file_2017 = "Resources/online-data-chapter-2-whr-2017.xlsx"
file_2016 = "Resources/Online-data-for-chapter-2-whr-2016.xlsx"
file_2015 = "Resources/Chapter2OnlineData_Expanded2015.xlsx"
```

```python
# Read our Data file with the pandas library
sheet1_2018_df = pd.read_excel(file_2018, 'Table2.1')
sheet2_2018_df = pd.read_excel(file_2018, 'Figure2.2')
sheet2_2016_df = pd.read_excel(file_2016, 'Figure2.2')
sheet2_2017_df = pd.read_excel(file_2017, 'Figure2.2')
sheet2_2015_df = pd.read_excel(file_2015, 'Figure2.2',skiprows=3)
```

```python
renamed_sheet1_2018 = sheet1_2018_df.rename(columns={"year":"Year","country":"Country"})
renamed_sheet1_2018.head()
```

```python
#remove columns we do not need
sheet2_2018_df.drop(sheet2_2018_df.columns.to_series()["Unnamed: 11":"Unnamed: 23"], axis=1, inplace=True)
sheet2_2018_df.head()
```

```python
# Using .rename(columns={}) in order to rename columns
renamed_2016 = sheet2_2016_df.rename(columns={"Dystopia (2.33) + residual":"Dystopia Residual"})
renamed_2016.head()
```

```python
renamed_2017 = sheet2_2017_df.rename(columns={"Dystopia (1.85) + residual":"Dystopia Residual"})
renamed_2017.head()
```

```python
renamed_2018 = sheet2_2018_df.rename(columns={"Dystopia (1.92) + residual":"Dystopia Residual"})
renamed_2018.head()
```

```python
#add year
renamed_2018["Year"] = 2018
renamed_2017["Year"] = 2017
renamed_2016["Year"] = 2016
sheet2_2015_df["Year"] = 2015
```

```python
#remove rows of data before 2016
cleansheet1_2018 = renamed_sheet1_2018[renamed_sheet1_2018.Year>=2016]
cleansheet1_2018.head()
```

```python
sheet2_2015_df.head()
```

```python
#create happiness score for 2015
sheet2_2015_df["Happiness score"] = sheet2_2015_df["Explained by: Log GDP per capita"]+sheet2_2015_df["Explained by: Social support"]+\
sheet2_2015_df["Explained by: Healthy life expectancy"]+sheet2_2015_df["Explained by: Freedom to make life choices"]+\
sheet2_2015_df["Explained by: Generosity"]+sheet2_2015_df["Explained by: Perceptions of corruption"]
sheet2_2015_df.head()
```

```python
#rename column Country
renamed_2015 = sheet2_2015_df.rename(columns={"country":"Country", "Explained by: Log GDP per capita":"Explained by: GDP per capita"})
renamed_2015.head()
```

```python
cleaned_2015 = renamed_2015[['Country', 'Happiness score', 
       'Explained by: GDP per capita', 'Explained by: Social support',
       'Explained by: Healthy life expectancy',
       'Explained by: Freedom to make life choices',
       'Explained by: Generosity', 'Explained by: Perceptions of corruption',
       'Year']]
cleaned_2015.head()
```

```python
#Concatenate
happiness_df = [renamed_2018,renamed_2017, renamed_2016, cleaned_2015]
concat_df = pd.concat(happiness_df)
concat_df
```

```python
cleaned_happiness_df = concat_df[['Country', 'Happiness score', 
       'Explained by: GDP per capita', 'Explained by: Social support',
       'Explained by: Healthy life expectancy',
       'Explained by: Freedom to make life choices',
       'Explained by: Generosity', 'Explained by: Perceptions of corruption',
       'Year']]
cleaned_happiness_df.head()
```

```python
# Combine the data into a single dataset
complete_df = pd.merge(cleaned_happiness_df, cleansheet1_2018, how="left", on=["Country", "Country","Year","Year"])
complete_df
```
